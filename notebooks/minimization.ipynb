```
import numpy as onp

import jax.numpy as np
from jax import random
from jax import jit

from jax_md import space, smap, energy, minimize, quantity, simulate
```

```
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
  
sns.set_style(style='white')

def format_plot(x, y):  
  plt.grid(True)
  plt.xlabel(x, fontsize=20)
  plt.ylabel(y, fontsize=20)
  
def finalize_plot(shape=(1, 1)):
  plt.gcf().set_size_inches(
    shape[0] * 1.5 * plt.gcf().get_size_inches()[1], 
    shape[1] * 1.5 * plt.gcf().get_size_inches()[1])
  plt.tight_layout()
```

# Harmonic Minimization

Here we demonstrate some simple example code showing how we might find the inherent structure for some initially random configuration of particles. Note that this code will work on CPU, GPU, or TPU out of the box.

First thing we need to do is set some parameters that define our simulation, including what kind of box we're using (specified using a metric function and a wrapping function).

```
N = 5000
dimension = 2
box_size = np.float64(80.0)
displacement, shift = space.periodic(box_size) 
```

Next we need to generate some random positions as well as particle sizes.

```
key = random.PRNGKey(0)
```

```
R = random.uniform(
  key, (N, dimension), minval=0.0, maxval=box_size, dtype=np.float32)

# The system ought to be a 50:50 mixture of two types of particles, one
# large and one small.
sigma = np.array([[1.0, 1.2], [1.2, 1.4]])
N_2 = int(N / 2)
species = np.where(np.arange(N) < N_2, 0, 1)
```

Then we need to construct our FIRE minimization function.

```
energy_fn = energy.soft_sphere_pair(
    displacement, species=species, sigma=sigma)
fire_init, fire_apply = minimize.fire_descent(energy_fn, shift)
fire_apply = jit(fire_apply)
fire_state = fire_init(R)
```

Now let's actually do minimization, keepting track of the energy as we go.

```
E = []
for i in range(200):
  fire_state = fire_apply(fire_state)
  
  E += [energy_fn(fire_state.position)]
R = fire_state.position
```

Let's plot the nearest distance for different species pairs. We see that particles on average have neighbors that are the right distance apart.

```
metric = lambda R: space.distance(space.map_product(displacement)(R, R))
dr = metric(R)
plt.plot(np.min(dr[:N_2, :N_2] + 5 * np.eye(N_2, N_2), axis=0), 'o', 
         label='$\\sigma_{AA}$')
plt.plot(np.min(dr[:N_2, N_2:] + 5 * np.eye(N_2, N_2), axis=0), 'o',
         label='$\\sigma_{AB}$')
plt.plot(np.min(dr[N_2:, N_2:] + 5 * np.eye(N_2, N_2), axis=0), 'o',
         label='$\\sigma_{BB}$')

plt.legend()
format_plot('', 'min neighbor distance')
finalize_plot()
```

Now let's plot the system. It's nice and minimized!

```
ms = 20
R_plt = onp.array(fire_state.position)

plt.plot(R_plt[:N_2, 0], R_plt[:N_2, 1], 'o', markersize=ms * 0.5)
plt.plot(R_plt[N_2:, 0], R_plt[N_2:, 1], 'o', markersize=ms * 0.7)

plt.xlim([0, np.max(R[:, 0])])
plt.ylim([0, np.max(R[:, 1])])

plt.axis('off')

finalize_plot((2, 2))
```

Finally, let's plot the energy trajectory that we observer during FIRE minimization.

```
plt.plot(E, linewidth=3)

format_plot('step', '$E$')
finalize_plot()
```

# Optimization via spatial partitioning

The above example uses a naive $O(N^2)$ algorithm to compute the force between each pair of particles at each step. However, in many cases particles only interact within some cutoff, $r_c$. In that case, an $O(N\log N)$ algorithm exists by using a spatial partition of the system into a grid whose cells have side length $r_c$. JAX, MD provides the function `smap.grid` to compute any function over a spatial partition in this way. We first compare the speed of execution for the naive implementation with the grid version.

```
R = fire_state.position
```

```
unoptimized_energy_fn = energy.soft_sphere_pair(
    displacement, species=species, sigma=sigma)
unoptimized_force_fn = jit(quantity.force(unoptimized_energy_fn))
```

```
energy_fn = energy.soft_sphere_pair(displacement, quantity.Dynamic, sigma=sigma)
force_fn = quantity.force(energy_fn)
force_fn = jit(smap.grid(force_fn, box_size, 1.4, R, species=species))
```

The first time the function is executed it must be jit compiled, therefore you will need to execute the following cells twice to get accurate timing information.

```
%%timeit

unoptimized_force_fn(R)
```

```
%%timeit

force_fn(R)
```

We observe around a 3.5x speedup. Next we can confirm that the two functions compute the same forces to within numerical precision.

```
dF = force_fn(R) - unoptimized_force_fn(R)
plt.semilogy(np.abs(np.reshape(dF, (-1,))), 'o')
format_plot('', '$\\Delta F$')
```

We can use the grid optimization to minimize a much larger system that we could have previously. Here we will minimize a 50k particle system.

```
N = 50000
dimension = 2
box_size = np.float32(252)
displacement, shift = space.periodic(box_size) 
```

As before, we need to generate some random positions as well as particle sizes.

```
R = random.uniform(
  key, (N, dimension), minval=0.0, maxval=box_size, dtype=np.float32)

# The system ought to be a 50:50 mixture of two types of particles, one
# large and one small.
sigma = np.array([[1.0, 1.2], [1.2, 1.4]], dtype=np.float32)
N_2 = int(N / 2)
species = np.where(np.arange(N) < N_2, 0, 1)
```

Now we construct a force function that operates on a grid.

```
energy_fn = energy.soft_sphere_pair(displacement, quantity.Dynamic, sigma=sigma)
force_fn = quantity.force(energy_fn)
cell_capacity = int(np.max(smap.count_cell_filling(R, box_size, 1.4)))
force_fn = jit(
    smap.grid(force_fn, box_size, 1.4, cell_capacity + 10, species=species))
```

Use the force function to construct FIRE descent dynamics as before. 

```
fire_init, fire_apply = minimize.fire_descent(force_fn, shift, quantity.Force)
fire_apply = jit(fire_apply)
fire_state = fire_init(R)
```

Let's actually do the minimization and draw the system.

```
for i in range(1000):
  fire_state = fire_apply(fire_state)
  
R = fire_state.position
```

```
ms = 6.3
R_plt = onp.array(fire_state.position)

plt.plot(R_plt[:N_2, 0], R_plt[:N_2, 1], 'o', markersize=ms * 0.5)
plt.plot(R_plt[N_2:, 0], R_plt[N_2:, 1], 'o', markersize=ms * 0.7)

plt.xlim([0, np.max(R[:, 0])])
plt.ylim([0, np.max(R[:, 1])])

plt.axis('off')

finalize_plot((2, 2))
```
